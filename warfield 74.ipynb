```python
#/Users/saching12/Desktop/python codes/ISM/
df = pd.read_csv(full_path, delim_whitespace=True, index_col=0)
name = " <h3>Structural Self-Interaction Matrix : </h3>"
display(Markdown(name))
display(df)
IPM = np.empty(df.shape, dtype=int)

size = len(df.columns)
i = 0
# index is not being used
for index, row in df.iterrows():
    size = len(df.columns)
    while size > 0:
        if i == size - 1:
            IPM[size - 1][i] = 1
        elif pd.isnull(row[str(size)]):
            break
        else:
            if row[str(size)] == 'V':
                IPM[i][size - 1] = 1
                IPM[size - 1][i] = 0
            elif row[str(size)] == 'A':
                IPM[i][size - 1] = 0
                IPM[size - 1][i] = 1
            elif row[str(size)] == 'X':
                IPM[i][size - 1] = 1
                IPM[size - 1][i] = 1
            elif row[str(size)] == 'O':
                IPM[i][size - 1] = 0
                IPM[size - 1][i] = 0
            else:
                print("Invalid Output")
        size -= 1
    i += 1




FRM = IPM.copy()
FRM = pd.DataFrame(data=FRM,columns=df.columns[::-1],index=df.index.values)
name = " <h3>Initial Reachability Matrix : </h3>"
display(Markdown(name))
FRM
```

```python
class Node: 
    def __init__(self,key): 
        self.U_1 = None
        self.F = None
        self.u_0 = key 
        self.V = None
        self.D = None
```

```python
def recursive_Subsystem(parameters,key,level,level_node):    
    display(Markdown('<h2>Level:'+str(level)+'</h2>'))

    print("Key:"+str(key))
    L = set()
    ## L-F
    U_1 = set()
    
    ## U_2 = S-L-key(a.k.a u_0)
    U_2 = set()
    
    ## Feedback set declaration
    F = set()
    V = set()
    D = set()

    ## creating lift set and Non lift set for key (a.k.a u_0)
    for d in parameters:
        if FRM.loc[key,d] == 1:
            L.add(d)
        else:
            U_2.add(d)
    print("Displaying L before F and U_2")
    print("Value of L is :",L)
    print("Value of U_2 is :",U_2)

    ## creating feedbackset for key(a.k.a u_0)
    for d in L:
        if FRM.loc[d,key] == 1:
            F.add(d)
    U_1 = L.difference(F)

    ## get vacancy set and drop set
    if len(U_2) > 1:
        for d in U_2:
            if FRM.loc[d,key] == 0:
                V.add(d)
            else :
                D.add(d)
    elif len(U_2) == 1:
        val = U_2.pop()
        if FRM.loc[val,key] == 0:
                V.add(val)
                D = None
        else:
            D.add(val)
            V = None
    else:
        V = None
        D = None

    print("L-F(U_1) : ",U_1)
    print("set F : ",F)
    print("Vacancy  Set :",V)
    print("Drop  Set",D)
    
    if level == 3:
        level_node.U_1 = Node(U_1) if len(U_1) == 1 else None
        level_node.F = F if len(F) >= 1 else None
        level_node.V = Node(V) if V is not None else None
        level_node.D = Node(D) if D is not None else None
    else:
        level_node.F = F
        if len(U_1) > 1:
            key = U_1.pop()
            print("u_0 of U_1:",key)
            new_node = Node(key)
            level_node.U_1 = recursive_Subsystem(U_1,key,level+1,new_node)
        else:
            print("U_1 else mode:")
            level_node.U_1 = Node(U_1)
        
        if V is not None and len(V) > 1 :
            key = V.pop()
            print("u_0 of Vacancy set: ",key)
            new_node = Node(key)        
            level_node.V = recursive_Subsystem(V,key,level+1,new_node)
        else:
            print("V else mode:")
            level_node.V = Node(V) if V is not None else None

        if D is not None and len(D) > 1:    
            key = D.pop()
            print("u_0 of Drop set: ",key)
            new_node = Node(key)        
            level_node.D = recursive_Subsystem(D,key,level+1,new_node)
        else:
            print("D else mode:")
            level_node.D = Node(D) if D is not None else None
    return level_node
```

```python
FRM.columns = FRM.columns.astype(int)
parameters = set(FRM.columns.to_list())
key = 12
parameters.remove(key)
print(key)
print(parameters)
Head_node = Node(key)
current_node = Head_node

final_node = recursive_Subsystem(parameters,key,1,current_node)
```

```python
def inorder_traversal(root,subset):
    if root:
        inorder_traversal(root.U_1,subset)
        if root.F is not None:
            subset.extend(list(root.F))
        if type(root.u_0) is int:
            subset.append(root.u_0)
        else:
            subset.extend(list(root.u_0))
        inorder_traversal(root.V,subset)
        inorder_traversal(root.D,subset)
final_parameter = []
inorder_traversal(final_node,final_parameter)       
```
